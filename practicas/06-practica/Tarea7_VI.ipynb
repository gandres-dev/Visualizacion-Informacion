Seleccionamos solamente la clase de automiles que nos interesan

```python
data = df[(df['class'] == 'suv') | (df['class'] == 'minivan') |  (df['class'] == 'compact')]
```

```python
p = sns.displot(data=data, x="cty", hue="class", kde=True, height=10); #  1040 x 800
p.fig.set_dpi(80)
plt.title('Curva de densidad con histogramas', fontsize=22);
```

```python
sns.displot(data=data, x="cty", hue="class", col="class", kde=True)
```

**Conclusines**

Vemos que el kilometraje es mayor con los autos de clase la suv, seguido despues de los compacto y por último minivan. Además vemos sigue distribucion normal, aunque sea muy pocos datos, vemos como inicia con un valor pequeño y va incrementadose hasta llegar a un maximo para posteriormente ir dismuyendos. Igual podemos observar que en nuestro conjunto de datos hay mayor cantidad de auto de la clase suv y muy pcos minivan. 

Tambien observamos que la importancia del colores, puede ayudarnos encontrar mejores patrones y darnos diferentes perséctivas de los datos que trabajos.
