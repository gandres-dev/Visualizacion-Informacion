```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftshift, ifft2

plt.style.use('seaborn-whitegrid') #classic
```

## Señal original

```python
T = 10  # Frecuencia
N = 256  # Sample rate

t = np.linspace(0, T, N)
y = 3 * np.sin(7.2 * np.pi * t + 2) + 1


plt.title("Señal original")
plt.xlabel('Time[s]')
plt.axis('equal')
plt.ylabel('Signal amplitude')
plt.plot(t, y, label='Señal', c='g')
```

## Transformada rapida de Fourier
### Espectro de frecuencia

```python
# Transformada rapida de Fourier
yt = fft(y)

# Graficando el espectro de Amplitud (Parte real)
#plt.plot(abs(yt))

# Vemos que tiene simetria par
#plt.plot(fftshift(abs(yt)))

plt.title("Espectro de frecuencia")
plt.xlabel('Frequency[s]')
plt.ylabel('Signal amplitude')
# Debemos de normalizar para tener esta forma la misma amplitud que la original
plt.plot(fftshift(abs(yt/N)))
```

### Espectro de fase

```python
# Tiene simetria impar, 
# Sacando la simetria ttf
# plt.plot(yt.imag)
# Graficando el espectro de Fase (Parte imaginaria)
plt.title("Espectro de fase")
plt.xlabel('Frequency[s]')
plt.ylabel('Signal amplitude')
# Debemos de normalizar para ten
plt.plot(fftshift(yt.imag))
```

## Actividad 2


```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
from scipy.fft import fft, ifft, fftshift, fft2, ifft2

[x,y] = np.meshgrid(np.linspace(0,10,64), np.linspace(0,10,64))

f = np.sin(2*pi*(0.2*x + 0.7*y))+np.random.uniform(0, 1, x.shape)
plt.pcolor(x,y,f)
plt.title("Señal")
```

```python
plt.set_cmap("plasma")
fig, axs = plt.subplots(1,2,figsize=(10,5))

axs[0].pcolor(f)
axs[0].set_title("Transformada inversa")


f2 = fft2(f)      
fts = fftshift(f2)
aux = np.abs(fts)

axs[1].imshow(aux)
axs[1].set_title("Transformada en 2D")
```
