Seleccionamos solamente la clase de automiles que nos interesan

```python
data = df[(df['class'] == 'suv') | (df['class'] == 'minivan') |  (df['class'] == 'compact')]
```

```python
data['cty'].hist(figsize=(10, 8))
plt.title('Histograma sobre el kilometraje en ciudad', fontsize=22)
```

```python
data['cty'].plot(kind="density", figsize=(10, 8))
```

```python
plt.figure(figsize=(10,8))
plt.xticks(    
)
sns.distplot(data['cty']);
plt.title('Curva de densidad sobre kilometraje en ciudad', fontsize=22)
```

```python
# p.fig.set_dpi(100)
p = sns.displot(data=data, x="cty", hue="class", kde=True, height=8); #
p.fig.set_dpi(80)
# plt.title('Curva de densidad sobre kilometraje en ciudad', fontsize=22);
```

```python
fig, ax = plt.subplots(figsize=(10,8), dpi=80)
sns.kdeplot(data=data,  x="cty", hue="class", ax=ax)
plt.title('Curva de densidad sobre kilometraje en ciudad', fontsize=22);
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.kdeplot(data=data,  x="cty", hue="class", multiple="stack", ax=ax)
```
